```python
import pandas as pd
```

```python
data=pd.read_csv('Housing.csv')
df=data.copy()
df.head()
```

```python
# Getting the number of rows and columns in a data set.

print("Number of Rows in a dataset: ",df.shape[0]) 
print("Number of columns in a dataset: ",df.shape[1])
```

```python
status = pd.get_dummies(data['furnishingstatus'])
status.head()
```

```python
bool_columns = status.select_dtypes(include='bool').columns
status[bool_columns] = status[bool_columns].astype('int64')

# Check the result
print(status)
print(status.dtypes)
```

```python
df = pd.concat([data, status], axis = 1)
df.head()
```

```python
df.drop(columns='furnishingstatus', axis=1, inplace=True)
```

```python
df.head()
```

```python
varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
df[varlist] = df[varlist].apply(binary_map)
```

```python
X=df.loc[:,df.columns!='price']

y=df['price']
```

```python
# Train Test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression, Ridge, Lasso, BayesianRidge, ElasticNet
from sklearn.metrics import explained_variance_score, r2_score
```

```python
# 1. Linear Regression

ols=LinearRegression()
ols.fit(X_train,y_train)
ols_yhat=ols.predict(X_test)
```

```python
# 2. Ridge regression

ridge=Ridge(alpha=0.5)
ridge.fit(X_train,y_train)
ridge_yhat=ridge.predict(X_test)
```

```python
# 3. Lasso Regression

lasso=Lasso(alpha=0.01)
lasso.fit(X_train,y_train)
lasso_yhat=lasso.predict(X_test)
```

```python
# 4. Bayesian 
bayesian=BayesianRidge()
bayesian.fit(X_train,y_train)
bayesian_yhat=bayesian.predict(X_test)
```
